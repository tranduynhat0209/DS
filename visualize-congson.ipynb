```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import kurtosis, skew



data_df = pd.read_csv("data-cleaned-v2/data-cleaned.csv", index_col=0)
```

```python
### Task a: Mỗi enum (hướng nhà, ban công, nội thất, giấy tờ, đường rộng, Boolean)  vẽ 1 biểu đồ thể hiện tất cả giá trị (mỗi enum làm 1 box plot)



### a.1: huong nha
direction_price_df = data_df[['Hướng nhà', 'Giá']]
direction_price_df['Giá'] = direction_price_df['Giá'] / 1000000
directions = direction_price_df['Hướng nhà'].unique()
maxLen =  len(directions)

data_hn = []
for i in range(0, maxLen):
    visual_directions = direction_price_df[direction_price_df['Hướng nhà'] == directions[i]]
    data_hn.append(visual_directions['Giá'])
data_hn.append(direction_price_df['Giá'])
labels = list(directions)
labels.append('Tất cả')
fig, ax = plt.subplots(1)
plt.title("Biểu đồ Boxplot giá theo hướng nhà")
ax.set_xticklabels(labels)
ax.set_ylim([0, 200000])
ax.boxplot(data_hn)
fig.set_figheight(10)
fig.set_figwidth(20)

plt.xlabel('Hướng nhà')
plt.ylabel('Giá (triệu đồng)')
plt.show()


```
